```python
import string
def checkCompanyID(data):

    totale=len(data)
    analizzati=0
    positivi=0
    setLetters=set(string.ascii_uppercase)

    for i in data:
        if i!=0:
            print(i)
            analizzati+=1
            if i[0] in setLetters and i[1] in setLetters:
               positivi+=1
               print('starts with letter')

            #sono tutti numeri
            else:
                matchingPr=True
                for char in i:
                    matchingPr=matchingPr and  char.isdigit()
                if matchingPr:
                        positivi+=1
    
    falsiPositivi=analizzati-positivi

    print('totale:',totale)
    print('analizzati:',analizzati)
    print('positivi:',positivi)
    print('falsi Positivi',falsiPositivi)
```

```python
def frequencyWord(data):
    totale=len(data)
    analizzati=0

    freq=dict()
    for i in data:
        if i!=0:
            if i not in freq:
                freq[i]=1
            
            else:
                freq[i]+=1
        analizzati+=1

    totale_con_freq=0
    for k in freq:
        if freq[k]>1:
            totale_con_freq+=1
            print(k,'->',freq[k])
    

    print('parole con >1 occorrenza', totale_con_freq)
    print('paroleDiverse da 0',len(freq.keys()))
    print('analizzati',analizzati)
    print('totale', totale)

```

```python
def analisiOfficeAddress(data):
    import re
    haRegex = re.compile(r'([Gg][Ii][Rr] 0[Aa]{2})|((([A-Za-z][0-9]{1,2})|(([A-Za-z][A-Ha-hJ-Yj-y][0-9]{1,2})|(([A-Za-z][0-9][A-Za-z])|([A-Za-z][A-Ha-hJ-Yj-y][0-9][A-Za-z]?))))\s?[0-9][A-Za-z]{2})')
    
    contaPositivi=0
    contaNegativi=0
    
    contaTotaliNotNull=0
    totale=len(data)
    for i in data:
        if i!=0:
            contaTotaliNotNull+=1
            if (' ' not in i):
                contaNegativi+=1
            else:
                words=i.split(' ')
                postalCode=words[len(words)-1]
                print(postalCode)
                matching=haRegex.match(postalCode)
                if matching:
                    contaPositivi+=1
                else:
                    contaNegativi+=1
        else:
            contaNegativi+=1

    
    print('Falsi Positivi:', contaNegativi)
    print('Positivi:',contaPositivi)
    print('Valori analizzati:',contaTotaliNotNull)
    print('totale:',totale)
```

```python
def check_shortDate(data):
    words=data.split('-')
    prefix=words[0]
    month=words[1]
    year=words[2]

    print('words ', words)
    print('prefix', prefix)
    print('year', year)
    print('month',month)

    matchingPr=True
    matchingMo=True
    mantchingY=True
    #prefix
    for char in prefix:
        print(char)
        matchingPr=matchingPr and char.isdigit()

    for y in year:
        print(y)
        mantchingY=mantchingY and y.isdigit()

    matchingMo=len(month)==3 or len(month)==4

    return matchingMo and matchingPr and mantchingY

def checkLongDate(data):
    words=data.split(' ')
    prefix=words[0]
    month=words[1]
    year=words[2]

    print('words ', words)
    print('prefix', prefix)
    print('year', year)
    print('month',month)

    matchingPr=True
    matchingMo=True
    mantchingY=True
    #prefix
    for char in prefix:
        print(char)
        matchingPr=matchingPr and char.isdigit()

    for y in year:
        print(y)
        mantchingY=mantchingY and y.isdigit()

    month=month.capitalize()
    monts=set(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
    if month in monts:
        matchingMo=True
    else:
        matchingMo=False

    return matchingMo and matchingPr and mantchingY


def checkDate(data):
    print(data)
    contaPositivi=0
    contaNegativi=0
    contaTotaliNotNull=0
    totale=len(data)
    
    for i in data:
        print(i)
        if i!=0 and type(i)==str:
            if '-' in i:
                match=check_shortDate(i)
            else:
                match=checkLongDate(i)

            print('Match:', match)
            if match==True:
                contaPositivi+=1
            else:
                contaNegativi+=1
            
            contaTotaliNotNull+=1
        else:
            contaNegativi+=1
    print('Falsi Positivi:', contaNegativi)
    print('Positivi:',contaPositivi)
    print('Valori analizzati:',contaTotaliNotNull)
    print('totale:',totale)
```

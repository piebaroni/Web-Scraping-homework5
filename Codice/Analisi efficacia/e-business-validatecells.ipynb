```python
def checkName(data):
    totale=len(data)
    analizzati=0
    positivi=0
    falsiPositivi=0
    negativi=[]
    examples=[]

    for i in data:
        if i!=0:
            analizzati+=1
            to_check=i[-10:]
            
            if '(' in to_check and ')' in to_check:
                code=to_check[1:len(to_check)-1]
                
                
                if code.isdigit():
                    positivi+=1
                    examples.append(i)
                else:
                    negativi.append(i)
                    falsiPositivi+=1
            else:
                falsiPositivi+=1
                negativi.append(i)


    print('Falsi Positivi:', falsiPositivi)
    print('Negativi:',negativi)
    print('Positivi:',positivi)
    print('Examples:',examples[0:4])
    print('Valori analizzati:',analizzati)
    print('totale:',totale)
```

```python
def analisiOfficeAddress(data):
    totale=len(data)
    analizzati=0
    positivi=0
    falsiPositivi=0
    negativi=[]
    examples=[]
    import re
    haRegex=re.compile(r'[0-9]{5}')
    for i in data:
        if i!=0:
            analizzati+=1
            
            i=''.join(i.split())    #elimino \t
            
            last5words=i[-5:]
           
            
            if haRegex.match(last5words):
                positivi+=1
                examples.append(i)
            else:
                falsiPositivi+=1
                negativi.append(i)


    print('Falsi Positivi:', falsiPositivi)
    print('negativi',negativi)
    print('Positivi:',positivi)
    print('Examples:',examples[0:4])
    print('Valori analizzati:',analizzati)
    print('totale:',totale)
        
```

```python
def checkDay(data):
    if not data.isdigit():
        return False
    data=int(data)   
    if data in range(1,32):
        return True
    return False

def chechMonth(data):
    if not data.isdigit():
        return False
    data=int(data)   
    if data in range(1,13):
        return True
    return False

def checkYear(data):
    if not data.isdigit():
        return False
    data=int(data)
    if data in range(1900,2025):
        return True
    return False

def checkDate(data):
    totale=len(data)
    positivi=0
    falsiPositivi=0
    analizzati=0
    negativi=[]
    examples=[]
    for i in data:
        if i!=0:
            
            analizzati+=1
            parts=i.split('.')
            
            correct=checkDay(parts[0]) and chechMonth(parts[1])and checkYear(parts[2])
            if correct==True:
                positivi+=1
                examples.append(i)
            else:
                falsiPositivi+=1
                negativi.append(i)

    print('Falsi Positivi:', falsiPositivi)
    print('Negativi:',negativi)
    print('Positivi:',positivi)
    print('Examples:',examples[0:4])
    print('Valori analizzati:',analizzati)
    print('totale:',totale)
    
```

```python
def frequencyWord(data):
    
    totale=len(data)
    analizzati=0

    freq=dict()
    for i in data:
        if i!=0:
            if i not in freq:
                freq[i]=1
            
            else:
                freq[i]+=1
        analizzati+=1

    totale_con_freq=0
    print('FREQUENZE')
    for k in freq:
        if freq[k]>1:
            totale_con_freq+=1
            print(k,'->',freq[k])
    

    
    print('Numero di valori distinti trovao:',len(freq.keys()))
    print('parole di valori con frequenza >1: ', totale_con_freq)
    print('analizzati: ',analizzati)
    print('totale: ', totale)

```

```python
def checkName(data):
    totale=len(data)
    analizzati=0
    positivi=0
    falsiPositivi=0

    for i in data:
        if i!=0:
            analizzati+=1
            to_check=i[-10:]
            print(to_check)
            if '(' in to_check and ')' in to_check:
                code=to_check[1:len(to_check)-1]
                print(code)
                
                if code.isdigit():
                    positivi+=1
                else:
                    falsiPositivi+=1
            else:
                falsiPositivi+=1


    print('Falsi Positivi:', falsiPositivi)
    print('Positivi:',positivi)
    print('Valori analizzati:',analizzati)
    print('totale:',totale)
```

```python
def analisiOfficeAddress(data):
    totale=len(data)
    analizzati=0
    positivi=0
    falsiPositivi=0
    import re
    haRegex=re.compile(r'[0-9]{5}')
    for i in data:
        if i!=0:
            analizzati+=1
            print(i)
            i=''.join(i.split())    #elimino \t
            print('After',i)
            last5words=i[-5:]
           
            print('last 5 letters:', last5words)
            if haRegex.match(last5words):
                positivi+=1
            else:
                falsiPositivi+=1


    print('Falsi Positivi:', falsiPositivi)
    print('Positivi:',positivi)
    print('Valori analizzati:',analizzati)
    print('totale:',totale)
        
```

```python
def checkDay(data):
    if not data.isdigit():
        return False
    data=int(data)   
    if data in range(1,32):
        return True
    return False

def chechMonth(data):
    if not data.isdigit():
        return False
    data=int(data)   
    if data in range(1,13):
        return True
    return False

def checkYear(data):
    if not data.isdigit():
        return False
    data=int(data)
    if data in range(1900,2025):
        return True
    return False

def checkDate(data):
    totale=len(data)
    positivi=0
    falsiPositivi=0
    analizzati=0
    for i in data:
        if i!=0:
            print(i)
            analizzati+=1
            parts=i.split('.')
            
            correct=checkDay(parts[0]) and chechMonth(parts[1])and checkYear(parts[2])
            if correct==True:
                positivi+=1
            else:
                falsiPositivi+=1

    print('Falsi Positivi:', falsiPositivi)
    print('Positivi:',positivi)
    print('Valori analizzati:',analizzati)
    print('totale:',totale)
    
```

```python
def frequencyWord(data):
    totale=len(data)
    analizzati=0

    freq=dict()
    for i in data:
        if i!=0:
            if i not in freq:
                freq[i]=1
            
            else:
                freq[i]+=1
        analizzati+=1

    totale_con_freq=0
    for k in freq:
        if freq[k]>1:
            totale_con_freq+=1
            print(k,'->',freq[k])
    

    print('parole con >1 occorrenza', totale_con_freq)
    print('paroleDiverse da 0',len(freq.keys()))
    print('analizzati',analizzati)
    print('totale', totale)

```

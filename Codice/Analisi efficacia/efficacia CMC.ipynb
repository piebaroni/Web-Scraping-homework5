```python
import pandas as pd
import numpy as np
import os
from itertools import combinations
import ast
import re
import matplotlib.pyplot as plt

color_of_edge= '#000000'
```

```python
df = pd.read_csv('C:\Web-Scraping-homework5\Profiling\DatasetParsed\cmc\cmc-aziende.csv', encoding='cp1252')
```

```python
df.head
```

## EFFICACIA XPATH CODICE

```python
code = df['Company code'].tolist()
```

```python
correct_code = 0
for elem in code:
    x = re.search("^(?:[A-Z]|[A-Z][A-Z]|[A-Z][A-Z][A-Z]|[A-Z][A-Z][A-Z][A-Z]|[A-Z][A-Z][A-Z][A-Z][A-Z])$", elem)
    if x:
        correct_code += 1
    else:
        print(elem)
d = {}
d['Aziende Totali'] = len(code)
d['Codici Conformi'] = correct_code
d['Codici Nulli'] = 0

names = list(d.keys())
values = list(d.values())
```

```python
df2 = pd.DataFrame(values, names)
print(df2)
ax = df2.plot.bar(color = '#c8a2c8', edgecolor=color_of_edge, linewidth=1)

ax.bar_label(ax.containers[0])
plt.bar(values, len(values))
plt.show()
```

## EFFICACIA XPATH MARKETCAP 

```python
mc = df['Marketcap'].tolist()

correct_mc = 0
for elem in mc:
    #x = re.search("^[0-9].+[.]0$", str(elem))
    x = re.search('^(?:[1-9]\d*|0)?(?:\.\d+)?$', str(elem))
    if x:
        correct_mc += 1
    else:
        print(elem)

print(correct_mc)
```

```python
d = {}
d['Aziende Totali'] = len(mc)
d['MarketCap Conformi'] = correct_mc
d['MarketCap Nulli'] = 1

names = list(d.keys())
values = list(d.values())

df2 = pd.DataFrame(values, names)
print(df2)
ax = df2.plot.bar(color = '#c8a2c8', edgecolor=color_of_edge, linewidth=1)

ax.bar_label(ax.containers[0])
plt.bar(values, len(values))
plt.show()
```

## EFFICACIA XPATH EARNINGS

```python
earnings = df['Earnings'].tolist()

correct_e = 0
for elem in earnings:
    #x = re.search("^[0-9].+[.]0$", str(elem))
    x = re.search('^[-](?:[1-9]\d*|0)?(?:\.\d+)?$', str(elem))
    if x:
        correct_e += 1
    else:
        y = re.search('^(?:[1-9]\d*|0)?(?:\.\d+)?$', str(elem))
        if y:
            correct_e += 1
        else:
            print(elem)

print(correct_e)
```

```python
d = {}
d['Aziende Totali'] = len(earnings)
d['Earnings Conformi'] = correct_e
d['Earnings Nulli'] = 4

names = list(d.keys())
values = list(d.values())

df2 = pd.DataFrame(values, names)
print(df2)
ax = df2.plot.bar(color = '#c8a2c8', edgecolor=color_of_edge, linewidth=1)

ax.bar_label(ax.containers[0])
plt.bar(values, len(values))
plt.show()
```

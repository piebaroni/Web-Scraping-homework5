```python
from bs4 import BeautifulSoup
import requests
from lxml import etree
```

```python
headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }
```

```python
def get_info(url):
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    dom = etree.HTML(str(soup))
    
    #Company name
    print(dom.xpath('//p[@class = "heading-xlarge"]')[0].text)
    
    #Company number
    print(dom.xpath('//p[@id = "company-number"]/strong')[0].text)
    
    #Company type
    type = dom.xpath('//dt[text() = "Company type"]/../dd')[0].text
    type = type.replace(' ', '')
    type = type.replace('\n', '')
    print(type)
    
    #Registered office address
    if(type == 'Overseasentity'):
        address = dom.xpath('//dt[text() = "Correspondence address"]/../dd')[0].text
        address = address.replace('\n', '')
        address = address.replace(' ', '')
        print(address)
    else:
        address = dom.xpath('//dt[text() = "Registered office address"]/../dd')[0].text
        address = address.replace(' ', '')
        address = address.replace('\n', '')
        print(address)
        #Nature of business
        nature = dom.xpath('//span[@id = "sic0"]')[0].text
        nature.replace(' ', '')
        nature.replace('\n', '')
        print(nature)
    
    #Company status
    status = dom.xpath('//dt[text() = "Company status"]/../dd')[0].text
    status = status.replace(' ', '')
    status = status.replace('\n', '')
    print(status)
    
    #Registration Date
    if(status == 'Registered'):
        print(dom.xpath('//dt[text() = "Registered on"]/../dd')[0].text)
    else:
        print(dom.xpath('//dt[text() = "Incorporated on"]/../dd')[0].text)
    
    #Dissolved on
    if(status == 'Dissolved'):
        print(dom.xpath('//dt[text() = "Dissolved on"]/../dd')[0].text)
```

```python
b = 0
url = 'https://find-and-update.company-information.service.gov.uk'    
for x in range(1, 200):
    if (b >= 1000): 
        break
    url2 = url + '/search/companies?q=a&page=' + str(x)
    req = requests.get(url2, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    for a in soup.find_all('a', class_ = 'govuk-link', href = True):
        if '/company' in a['href']:
            b += 1
            get_info(url + a['href'])
print(b)
```
